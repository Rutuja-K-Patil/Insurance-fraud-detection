# Data Cleaning and Preprocessing

 Find zero values

```python
#Count the zero values from Dataframe

def columns_with_zeros(df):
    zero_counts = df.eq(0).sum()
    for column_name, count in zero_counts.items():
        if count > 0:
            print(f"Column '{column_name}' has zero's : {count} ")

columns_with_zeros(df)
```

 Check for Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_and_find_outliers(df):
    for column in df:
        if df[column].dtype in ['int64', 'float64']:
            plt.figure(figsize=(8, 6))
            sns.boxplot(x=df[column])

            plt.title(f'Boxplot of {column}')
            plt.xlabel(column)

            plt.show()

            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]

            print(f'Outliers in {column}:')
            if len(outliers) == 0:
                print('No outliers found.')
            else:
                print(outliers)

visualize_and_find_outliers(df)
```

# one hot encoding

```python
# Define cat_cols as a list of categorical column names
cat_columns = ['Plan','Policy Term','Mode']
df_encoded = pd.get_dummies (data=df , columns =cat_columns)
```

```python
df_encoded.shape
```

# Split the dataframe into x and y

```python
X = df_encoded.drop(columns=['Anomaly Flag'])
y = df_encoded['Anomaly Flag']
```

# Balancing the dataset

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state = 42)
X, y = smote.fit_resample(X, y)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X,y,test_size=0.2,random_state = 7)
```

# Build Model

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(
    n_estimators = 1050,
    max_depth = 10,
    min_samples_split = 10,
    min_samples_leaf = 4,
)
classifier.fit(X_train,Y_train)
```

```python
classifier_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Train Accuracy of RandomForestClassifier :",classifier.score(X_train,Y_train)*100)
print("Test Accuracy of RandomForestClassifier :",classifier.score(X_test,Y_test)*100)
print("Accuracy Score of RandomForestClassifier :",accuracy_score(Y_test,classifier_predict)*100)
```

```python
#Confusion Matrix and Classifiaction Report of Random Forest Classifier

from sklearn.metrics import classification_report,confusion_matrix
cm = confusion_matrix(Y_test,classifier_predict)
sns.heatmap(confusion_matrix(Y_test,classifier_predict),annot=True,fmt='d')
```

```python
print("Classification Report of Random Forest Classifier :\n ",classification_report(Y_test,classifier_predict))
```

# Hyper Parameter Tunning For Random Forest

```python
#MAnual Hyperparameter Tuning

model=RandomForestClassifier(n_estimators=100,min_samples_leaf=1,
                             random_state=100,min_samples_split=2)

model.fit(X_train,Y_train)
predictions=model.predict(X_test)
print(confusion_matrix(Y_test,classifier_predict))
print(accuracy_score(Y_test,classifier_predict))
print(classification_report(Y_test,classifier_predict))
```
